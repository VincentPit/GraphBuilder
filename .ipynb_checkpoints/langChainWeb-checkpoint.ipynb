```python
from langchain.document_loaders import WebBaseLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.llms import OpenAI
from langchain.chains import RetrievalQA
from langchain.callbacks import get_openai_callback
 
loader = WebBaseLoader(web_paths=["https://www.dfrobot.com.cn", "https://makelog.dfrobot.com.cn/"])
documents = loader.load()



text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
texts = text_splitter.split_documents(documents)
 

```

```python
embeddings = OpenAIEmbeddings()
docsearch = Chroma.from_documents(texts, embeddings)

llm = OpenAI(temperature=0)
retriever = docsearch.as_retriever()
qa = RetrievalQA.from_chain_type(llm, chain_type="stuff", retriever=retriever)
 
query = "What is the content of the webpage?"
with get_openai_callback() as cb:
    useful_response = qa({"query": query})
    print(useful_response["result"])
    print(f"Token usage: {cb}")  # 显示 token 使用情况
```

```python
import asyncio
import aiohttp
from bs4 import BeautifulSoup
from urllib.parse import urljoin

visited = set()

async def extract_links(url):

    
    try:
        async with aiohttp.ClientSession() as session:
            async with session.get(url) as response:
                html = await response.text()
                soup = BeautifulSoup(html, 'html.parser')
                return [urljoin(url, a.get('href')) for a in soup.find_all('a') if a.get('href')]
    except Exception as e:
        print(f"Error fetching {url}: {e}")
        return []

async def recursive_crawl(url, delay=1):  # 添加一个延迟参数
    if url in visited:
        return
    
    visited.add(url)
    print(f"Processing: {url}")
    links = await extract_links(url)
    for link in links:
        await asyncio.sleep(delay)  # 在每个请求之前添加延迟
        await recursive_crawl(link)
```

```python
# 示例调用
start_url = "https://www.dfrobot.com.cn/"
asyncio.run(recursive_crawl(start_url))


with open('links.txt', 'w') as f:
    for link in visited:
        f.write(link + '\n')
```
